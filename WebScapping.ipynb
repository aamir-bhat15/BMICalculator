```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.scrapethissite.com/pages/forms/"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.find_all('p', class_='lead')
```

```python
title = soup.find_all('th')
title_header = [titleh.text.strip() for titleh in title]
print(title_header)
```

```python
soup.find('p', class_='lead').text.strip()
```

```python
import pandas as pd
df = pd.DataFrame(columns=title_header)
```

```python
df
```

```python
data = soup.find_all('tr')
```

```python
for row in data[1:]:
    row_data = row.find_all('td')
    ind_row_data = [data.text.strip() for data in row_data ]
    
    length=len(df)
    df.loc[length] = ind_row_data
```
